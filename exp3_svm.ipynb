```python
import pandas as pd 
```

```python
data=pd.read_csv('svm.csv')
```

```python
data.columns
```

```python
data.columns=data.columns.str.strip()
```

```python
data.columns
```

```python
data.columns = ['IP Mean', 'IP Sd', 'IP Kurtosis', 'IP Skewness', 
              'DM-SNR Mean', 'DM-SNR Sd', 'DM-SNR Kurtosis', 'DM-SNR Skewness', 'target_class']
```

```python
data.columns
```

```python
data['target_class'].value_counts()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['IP Kurtosis']=data['IP Kurtosis'].fillna(data['IP Kurtosis'].median())
data['DM-SNR Sd']=data['DM-SNR Sd'].fillna(data['DM-SNR Sd'].median())
data['DM-SNR Skewness']=data['DM-SNR Skewness'].fillna(data['DM-SNR Skewness'].median())
```

```python
data.isnull().sum()
```

```python
X=data.drop(['target_class'],axis=1)
Y=data['target_class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.2,random_state=0)
```

```python
cols=x_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train=pd.DataFrame(x_train, columns=[cols])
x_test=pd.DataFrame(x_test, columns=[cols])
```

```python
x_train.describe()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc=SVC()
```

```python
svc.fit(x_train, y_train)
```

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
linear=SVC(kernel='linear',C=1.0)
linear.fit(x_train, y_train)
y_pred=linear.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
